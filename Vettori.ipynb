Vettori
Crea una classe che rappresenti un vettore numerico. La classe accetta in input una lista di numeri che saranno i valori del vettore. La classe deve avere queste funzionalità:

Stampando un'oggetto Vector il risultato sarà la lista dei suoi valori.
Sommando due oggetti Vector il risultato sarà un nuovo oggetto Vector composto dalla somma dei valori dei due vettori alle posizioni corrispondenti (es. Vector([1,2,3]) + Vector([2,4,6]) = Vector([3,6,9]).
Sottraendo un'oggetto Vector ad un'altro oggetto Vector il risultato sarà un nuovo oggetto Vector composto dalla differenza dei valori dei due vettori alle posizioni corrispondenti Vector([2,4,6]) + Vector([1,2,3]) = Vector([1,2,3]).
Moltiplicando due oggetti Vector il risultato dovrà essere il prodotto scalare dei due vettori, il prodotto scalare è definito come la somma del prodotto dei singoli elementi corrispondenti dei due vettori (es. Vector([1,2,3]) + Vector([2,4,6]) = 28.
Per i punti 2,3 e 4 i vettori devono avere uguale dimensione, in caso di dimensioni differenti stampa "I vettori hanno dimensione differente" e ritorna None.

Inoltre la classe deve avere i seguenti metodi:

.sum(): ritorna la somma di tutti gli elementi del vettore.
.norm(): ritorna la norma del vettore, cioè la radice quadrata del prodotto scalare tra il vettore e se stesso (es. Vector([10, 20, 30, 40]).norm() = 54.77)
Inoltre la classe deve supportare l'indexing

es.
v = Vector([2, 4, 6])
print(v[1]) # stampa 4

```python
class Vector():
    
    def __init__(self, values):
        self._values = values
        
    def _is_len_eq(self,vector):
        
        if len(self._values) != len(vector._values):
            print("I due vettori hanno dimensioni diverse")
            return False
        return True
        
    def __getitem__(self, i):
        return self._values[i]
    
    def __repr__(self):
        return str(self._values)
    
    def __add__(self, vector):
        
        if not self._is_len_eq(vector):
            return None
        
        list_sum = []
        
        for val1, val2 in zip(self._values, vector._values):
            list_sum.append(val1+val2)
            
        return Vector(list_sum)
    
    def __sub__(self, vector):
        
        if not self._is_len_eq(vector):
            return None
        
        list_diff = []
        
        for val1, val2 in zip(self._values, vector._values):
            list_diff.append(val1-val2)
            
        return Vector(list_diff)
        
        
    
    def __eq__(self, vector):
        
        if not self._is_len_eq(vector):
            return None
        
        for val1, val2 in zip(self._values, vector._values):
            if val1!=val2:
                return False
        
        return True
        
    
    def __mul__(self, vector):
        return self.dot(vector)
    
    
    def dot(self, vector):
        
        if not self._is_len_eq(vector):
            return None
        
        prod = 0
        
        for val1, val2 in zip(self._values, vector._values):
            prod+=val1*val2
            
        return prod
        
        
    
    def sum(self):
        return sum(self._values)
    
    def norm(self):
        return self.dot(self)**(1/2)
    
    
```

```python
v1 = Vector([1, 2, 3])
print(v1)
print(v1[0])
print(v1.sum())

v2 = Vector([2, 4, 6])
print(v1 + v2)
print(v2 - v1)

print(v1==v1)
print(v1==v2)

print(v1*v2)

v3 = Vector([10, 20, 30, 40])
print(v1*v3)

print(v3.norm())
```
